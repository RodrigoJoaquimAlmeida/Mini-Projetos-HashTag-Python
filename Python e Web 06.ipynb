# Desafio - Rotina de Baixar uma Planilha da Web

- Imagine que você trabalhe no Mercado Financeiro e tem que todo dia/semana baixar uma planilha com as cotações do dólar
- Usaremos o site investing.com para baixar esses dados
- O link onde ficam esses dados é: https://br.investing.com/currencies/usd-brl-historical-data
- Escolhemos o site investing.com porque ele é cheio de coisinha chata que vai obrigar a gente a fazer um código completo
- Crie uma conta no site antes de começar, é gratuito

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
import time
import pandas as pd

options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')

navegador = webdriver.Chrome(options=options)
```

```python
navegador.get('https://br.investing.com/currencies/usd-brl-historical-data')
time.sleep(40)
navegador.find_element(By.ID, 'onetrust-accept-btn-handler').click()
time.sleep(10)
baixar_dados = navegador.find_element(By.XPATH, '//*[@id="column-content"]/div[4]/div/a')
ActionChains(navegador).click(baixar_dados).perform()
time.sleep(5)
navegador.find_element(By.ID, 'loginFormUser_email').send_keys('xxxxxxx@gmail.com')
time.sleep(1)
navegador.find_element(By.ID, 'loginForm_password').send_keys('xxxxxxx')
time.sleep(1)
clicar_singup = navegador.find_element(By.XPATH, '//*[@id="signup"]/a')
ActionChains(navegador).click(clicar_singup).perform()
time.sleep(5)
baixar_dados = navegador.find_element(By.XPATH, '//*[@id="column-content"]/div[4]/div/a')
ActionChains(navegador).click(baixar_dados).perform()
time.sleep(2)
cotacao_df = pd.read_csv('USD_BRL Dados Históricos.csv')
display(cotacao_df)

```
